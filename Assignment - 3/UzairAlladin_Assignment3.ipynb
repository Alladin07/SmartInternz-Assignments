#### Problem Description:

House price prediction is a common problem in the real estate industry and involves predicting the selling price of a house based on various features and attributes. The problem is typically approached as a regression problem, where the target variable is the price of the house, and the features are various attributes of the house

The features used in house price prediction can include both quantitative and categorical variables, such as the number of bedrooms, house area, bedrooms, furnished, nearness to main road, and various amenities such as a garage and other factors that may influence the value of the property.

Accurate predictions can help agents and appraisers price homes correctly, while homeowners can use the predictions to set a reasonable asking price for their properties. Accurate house price prediction can also be useful for buyers who are looking to make informed decisions about purchasing a property and obtaining a fair price for their investment.

##### Question 5 - Find the outliers and replace the outliers

```python
df.boxplot(column = 'area')
```

```python
#Handling the Outliers of Area
area_median = df['area'].median()

#Calculating the upper and the lower threshold
area_upper_threshold = df['area'].quantile(0.75) + 1.5 * (df['area'].quantile(0.75) - df['area'].quantile(0.25))
area_lower_threshold = df['area'].quantile(0.25) - 1.5 * (df['area'].quantile(0.75) - df['area'].quantile(0.25))

#Changing the Outliers
df.loc[df['area'] > area_upper_threshold] = area_median
df.loc[df['area'] < area_lower_threshold] = area_median
```

##### Question 7 - Split the data into dependent and independent variables.

```python
X = df.drop('price',axis=1)
X
```

```python
y = df['price']
y
```

##### Question 8 - Scale the independent variables

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
normalise = scale.fit_transform(X)
X = pd.DataFrame(normalise, columns=X.columns)
X.head(20)
```

##### Question 9 - Split the data into training and testing

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.info()
```

```python
X_test.info()
```

##### Question 10 - Build the Model

```python
from sklearn.linear_model import LinearRegression
linmodel = LinearRegression()
```

##### Question 11 - Train the Model

```python
linmodel.fit(X_train,y_train)
```

```python
print(linmodel.intercept_)
```

```python
coeff_df = pd.DataFrame(data=linmodel.coef_,index=X.columns,columns=['Coefficient'])
coeff_df
```

##### Question 12 - Test the Model

```python
predictions = linmodel.predict(X_test)
predictions
```

##### Question 13 - Measure the performance using Metrics.

```python
from sklearn import metrics
print('Mean Absolute Error (MAE): ', metrics.mean_absolute_error(y_test,predictions))
print('Mean Squared Error (MSE): ', metrics.mean_squared_error(y_test,predictions))
print('Root Mean Squared Error (RMSE): ', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```
